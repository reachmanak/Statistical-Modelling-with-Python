# Database

Put all your results in an SQLite3 database (remember, SQLite stores its databases as files in your local machine - make sure to create your database in your project's data/ directory!)

```python
# Import of sqllite package
import sqlite3
# Creating a connection and a database
conn = sqlite3.connect('London_database')
c = conn.cursor()
# Creating new table with defining data type
c.execute('CREATE TABLE IF NOT EXISTS londonbikes (IndexNum INTEGER PRIMARY KEY, name TEXT NOT NULL, distance INTEGER NOT NULL, restaurants TEXT, popularity INTEGER,rating INTEGER,address TEXT,Available_Bikes INTEGER NOT NULL,Bike_Stations INTEGER,Reviews INTEGER,Score INTEGER,empty_slots INTEGER)')
conn.commit()
Final = pd.read_csv('Final_data.csv', index_col=0)
Final.rename(columns = {'index':'IndexNum'},inplace=True)
#Structuring and selecting the columns for conversion
df = pd.DataFrame(Final, columns= ['IndexNum','Restaurants','distance','popularity','rating','address','Available.Bikes','Bike.Stations','review','empty_slots'])
```

```python
df.to_sql('londonbikes', conn, if_exists='replace', index = False)
```
